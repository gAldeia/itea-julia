```python
import re

import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline

from statannotations.Annotator import Annotator

sns.set_theme(style='white', font_scale=0.8)

plt.rcParams['text.usetex'] = True 
plt.rcParams['text.latex.preamble'] = r'\usepackage{sansmath} \sansmath'
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = 'Times New Roman'

plt.rc('font',   size=14)
plt.rc('axes',   titlesize=14) 
plt.rc('axes',   labelsize=14)
plt.rc('xtick',  labelsize=14)
plt.rc('ytick',  labelsize=14)
plt.rc('legend', fontsize=14)  
plt.rc('figure', titlesize=14)

pallete = sns.color_palette("colorblind", 9)

results_files = [
    '../results/results_ordinary_least_squares_adj.csv',
    '../results/results_ordinary_least_squares_adj_unrestricted.csv',
    #'../results/results_nothing.csv',
]

def solve_method_name_from_filename(text):
    rep = {"../results/results_": "", "_adj": "", ".csv": "",
           "ordinary_least_squares": "OLS",
           "levenberg_marquardt" : "LM",
           "_unrestricted" : " (U)",
           "nothing" : "No adjustment"}

    rep = dict((re.escape(k), v) for k, v in rep.items()) 
    
    pattern = re.compile("|".join(rep.keys()))
    
    return pattern.sub(lambda m: rep[re.escape(m.group(0))], text).replace("_", "+")
```

```python
df_results = []
for rf in results_files:
    adj_name = solve_method_name_from_filename(rf)
    
    df = pd.read_csv(rf)
    df['Adj_method'] = solve_method_name_from_filename(adj_name)

    df_results.append(df)
    
    print(f'- {adj_name} smallest expression: '+
          df.iloc[(df[df['Dataset']=='airfoil'])['N_nodes'].idxmin()]['Expr'] + '\n')
    
df_results = pd.concat(df_results, axis=0)

display(df_results)
```

```python
pairs =(
    [('airfoil', 'LM'), ('airfoil', 'LM_OLS')],
    [('airfoil', 'OLS'), ('airfoil', 'LM_OLS')],
    [('airfoil', 'LM_OLS'), ('airfoil', 'LM')],
)

for info in ['NMSE_train', 'NMSE_test', 'N_nodes', 'MAE_test', 'Exec_time']:
    
    df_results_pivoted = pd.pivot_table(df_results, values=info, index=['Dataset', 'Fold', 'Rep'],
                    columns=['Adj_method'])
    
    medians = df_results_pivoted.groupby('Dataset').quantile(0.5)
    q1 = df_results_pivoted.groupby('Dataset').quantile(0.25)
    q3 = df_results_pivoted.groupby('Dataset').quantile(0.75)
    iqr = q3.subtract(q1)
    
    merged = pd.DataFrame()

    float_format = lambda x: '%10.3f' % x

    for col_name in medians.columns:
        merged.loc[:, col_name] = ' $' + medians[col_name].apply(float_format) + \
                                  r' \pm ' + iqr[col_name].apply(float_format) + '$ '
    display(merged)    
    print(merged.to_latex(escape=False).replace('  ', ''))

    display(df_results.groupby(['Adj_method', 'Dataset'])['NMSE_test'].median().unstack(level=1).rank().mean(axis=1).sort_values(ascending=True))
    # Ordenando pela mÃ©dia
    meds = df_results.groupby(['Adj_method', 'Dataset'])['NMSE_test'].median().unstack(level=1).rank().mean(axis=1).sort_values(ascending=True)
    methods = meds.index.values #df_results['Adj_method'].unique()
    indexes = range(len(methods))
    #print(methods)
    pairs = list(
        [((d, methods[0]), (d, methods[i]))for i in range(1, len(methods))
        for d in df_results['Dataset'].unique()]
    )
    plot_params = {
        'data':      df_results,
        'x':         'Dataset',
        'y':         info,
        #"order":     states_order,
        "hue":       "Adj_method",
        "hue_order": meds.index.values,
        "palette":   {'OLS': pallete[3], 'OLS (U)': pallete[4], 'No adjustment':pallete[2]}#"viridis"
    }
    
    print(info)
    fig, ax = plt.subplots(1, 1, figsize=(6, 4))

    #meds = df_results.groupby('Adj_method').mean().sort_values(by=info, ascending=True)
    #print(meds.index.values)

    g = sns.boxplot(**plot_params, 
        showfliers=False, ax=ax, orient='v', notch=True, linewidth=1)
    #g.set_yscale("log")

    for spine in ['right', 'top']:
        ax.spines[spine].set_visible(False)

    ax.set_axisbelow(True)
    ax.grid(linestyle='dashed') #axis='y')

    # Add annotations
    annotator = Annotator(ax, pairs, **plot_params)
    annotator.configure(test="Mann-Whitney", verbose=False, text_format='star', line_width=1,
                       comparisons_correction='Holm-Bonferroni', line_offset=0.5,
                       correction_format='replace')
    _, results = annotator.apply_and_annotate()

    # Label and show

    #ax.legend(loc=(1.0, 0.5))

    box = ax.get_position()
    ax.set_position([box.x0, box.y0 + box.height * 0.1,
                     box.width, box.height * 0.9])

    # Put a legend below current axis
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.2),
              fancybox=True, shadow=True, ncol=5)

    ax.set_xlabel('')
    ax.set_ylabel(info.replace('_', ' '))

    plt.xticks(rotation=15, ha='right') 
    plt.tight_layout()
    plt.savefig(f'./plots/boxplot_{info}_OLS.pdf', bbox_inches='tight')
    plt.show()
```

```python
for info1, info2 in [('Exec_time', 'NMSE_test'), ('N_nodes', 'NMSE_test'), ('Exec_time', 'N_nodes'), ('NMSE_train', 'NMSE_test')]:
    fig, ax = plt.subplots(1, 1, figsize=(6, 4))
    
    sns.scatterplot(**{
            'data':      df_results[df_results['Dataset']=='airfoil'],
            'x':         info1,
            'y':         info2,
            "hue":       "Adj_method",
            "hue_order": meds.index.values,
            "palette":   {'OLS': pallete[3], 'OLS (U)': pallete[4], 'No adjustment':pallete[2]}#"viridis"
    }, ax=ax, style="Adj_method", alpha=0.75, s=100, markers=(['o', 's', 'd', 'X', '+'])[:len(results_files)])

    for spine in ['right', 'top']:
        ax.spines[spine].set_visible(False)

    ax.set_xlabel(info1.replace('_', ' '))
    ax.set_ylabel(info2.replace('_', ' '))
    
    ax.set_axisbelow(True)
    ax.grid(linestyle='dashed')

    ax.get_legend().set_title("")
    ax.legend(loc='upper right')
    
    #plt.xticks(rotation=15, ha='center') 
    plt.tight_layout()
    plt.savefig(f'./plots/scatter_{info1, info2}_OLS.pdf', bbox_inches='tight')
    plt.show()
```
